```python
import pandas as pd
```

```python
df=pd.read_csv("BlackFriday.csv")
del df["Product_Category_2"]
del df["Product_Category_3"]
df.head()
```

```python
df.groupby("Age").size().plot(kind="bar",title="Purchase Distribution by age")
```

```python
df["Age"].unique()
```

```python
for i in df["Age"].unique():
    print(f"{i}:{len(df[df['Age']==i])}")
```

```python
for i in df["Age"].unique():
    print(f"{i}:{df[df['Age']==i]['Product_ID'].nunique()}")
```

```python
lst=[]
for i in df["Age"].unique():
    lst.append([i,df[df['Age']==i]['Product_ID'].nunique()])
df_age=pd.DataFrame(lst,columns=["Age","Products"])
```

```python
df_age
```

```python
df_age.plot.bar(x="Age")
```

```python
df.groupby("Age")["Purchase"].sum().plot(kind="bar",title="Amount spent by age on purchase")
```

```python
df.groupby("Age")["Purchase"].mean().plot(kind="bar",title="Average Amount spent by age on purchase")
```

```python
df.groupby("Age")["Purchase"].mean()
```

```python
df.groupby("Marital_Status").size().plot(kind="pie",autopct="%.1f")
```

```python
df.groupby("Gender").size().plot(kind="pie",autopct="%.1f")
```
