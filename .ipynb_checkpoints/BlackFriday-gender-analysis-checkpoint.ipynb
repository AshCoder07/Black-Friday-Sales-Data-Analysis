```python
import pandas as pd
```

```python
df=pd.read_csv("BlackFriday.csv")

del df["Product_Category_2"]
del df["Product_Category_3"]
```

```python
df.head()
```

```python
for column in df.columns:
    print(f"{column}:{df[column].nunique()}")
```

```python
data=pd.DataFrame({"Ratio":[len(df[df["Gender"]=='M']),len(df[df["Gender"]=="F"])]},index=["Male","Female"])
```

```python
data
```

```python
data.plot(kind="pie",autopct="%.2f",y="Ratio")
```

```python
df.groupby("Gender").size().plot(kind="pie",autopct="%.2f",title="Gender Ratio")
```

```python
df.groupby("Gender")["Purchase"].sum()
```

```python
df.groupby("Gender")["Purchase"].sum().plot(kind="pie",autopct="%.2f",title="Ratio of purchase by gender")
```

```python
df.groupby("Gender")["Purchase"].mean().plot(kind="pie",autopct="%.2f",title="Average purchase ratio by gender")
```
